С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.
Создайте массив Numpy под названием f, содержащий косинусы элементов массива t.
Постройте линейную диаграмму, используя массив t для координат по горизонтали,а массив f - для координат по вертикали. Линия графика должна быть зеленого цвета.
Выведите название диаграммы - 'График f(t)'. Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'.
Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.


```python
t = np.linspace(0,10,51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t, f, color = 'green')
plt.show()
```

Выведите название диаграммы - 'График f(t)'. 
Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'.

```python
%config InlineBackend.figure_format = 'svg'
```

```python
plt.plot(t, f, color = 'green')
title_dict = {'fontsize':16, 'fontweight':'bold', 'color':'DarkRed'}
label_dict = {'fontsize':12,  'color':'#808080'}
plt.title('График f(t)', fontdict = title_dict)
plt.xlabel('Значения t', fontdict = label_dict)
plt.ylabel('Значения f', fontdict = label_dict)
plt.grid()
plt.show()
```

Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.

```python
plt.axis([0.5, 9.5, -2.5, 2.5 ])
plt.plot(t, f, color = 'green')
title_dict = {'fontsize':16, 'fontweight':'bold', 'color':'DarkRed'}
label_dict = {'fontsize':12,  'color':'#808080'}
plt.title('График f(t)', fontdict = title_dict)
plt.xlabel('Значения t', fontdict = label_dict)
plt.ylabel('Значения f', fontdict = label_dict)
plt.grid()
plt.show()
```

Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure 
с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных 
графика в сетке, состоящей из двух строк и двух столбцов. В каждом графике массив x используется 
для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте y1,
в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.Дайте название графикам:
    'График y1', 'График y2' и т.д.


```python
%config InlineBackend.figure_format = 'svg'
```

```python
plt.axis([-3,3, -10, 10])
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
```

```python
fig, ax = plt.subplots()
ax1 = ax
ax1.plot(x,y1)
```

```python
fig, ax = plt.subplots()
ax2 = ax
ax2.plot(x,y2)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax1,ax2, ax3, ax4 = ax.flatten()
ax1.plot(x,y1)
ax2.plot(x,y2)
ax3.plot(x,y3)
ax4.plot(x,y4)
fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace = 0.3)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax1,ax2, ax3, ax4 = ax.flatten()
ax1.plot(x,y1)
ax1.set_title('График y1')
ax1.set_xlim([-5,5])

ax2.plot(x,y2)
ax2.set_title('График y2')

ax3.plot(x,y3)
ax3.set_title('График y3')

ax4.plot(x,y4)
ax4.set_title('График y4')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace = 0.3)
```
